```python
# ==========================================
# ADVANCED INDUSTRIAL HR DATASET GENERATOR
# ==========================================

import pandas as pd
import numpy as np
import random

np.random.seed(42)

# Expanded realistic values
names = ["Aman","Priya","Rahul","Neha","Arjun","Sara","John","Emily","David","Meera","Karan","Anjali"]
countries = ["India","USA","UK","Canada","Australia"]
states = {
    "India": ["Punjab","Delhi","Haryana","Maharashtra","Gujarat","Karnataka","Tamil Nadu"],
    "USA": ["California","Texas","New York"],
    "UK": ["London","Manchester"],
    "Canada": ["Ontario","Quebec"],
    "Australia": ["Sydney","Melbourne"]
}
education = ["UG","PG","Diploma"]
genders = ["Male","Female"]

data = []

for i in range(1000):   # 🔥 1000 rows

    country = random.choice(countries)
    state = random.choice(states[country])

    exp = np.random.randint(0,10)
    skill = np.random.randint(5,10)
    comm = np.random.randint(5,10)
    tech = np.random.randint(5,10)
    interview = np.random.randint(5,10)

    certifications = np.random.randint(0,6)
    company_rating = np.random.randint(1,6)

    # 🎯 SMART INDUSTRIAL LOGIC (NOT RANDOM)
    score = (
        (skill * 0.25) +
        (comm * 0.15) +
        (tech * 0.30) +
        (interview * 0.20) +
        (certifications * 0.05) +
        (company_rating * 0.05)
    )

    if score > 7.5 and exp >= 2:
        result = 1
    else:
        result = 0

    salary = 30000 + (exp * 8000) + (skill * 2000)

    data.append([
        f"C{i+1:04}",
        random.choice(names),
        np.random.randint(21,36),
        random.choice(genders),
        country,
        state,
        random.choice(education),
        exp,
        skill,
        comm,
        tech,
        interview,
        certifications,
        company_rating,
        salary,
        random.choice([15,30,45,60,90]),
        result
    ])

df = pd.DataFrame(data, columns=[
    "Candidate_ID","Name","Age","Gender","Country","State","Education",
    "Experience","Skill_Score","Communication_Score","Technical_Test_Score",
    "Interview_Score","Certifications_Count","Previous_Company_Rating",
    "Expected_Salary","Notice_Period","Result"
])

# Save dataset
df.to_csv("advanced_hr_recruitment_data.csv", index=False)

print("✅ Dataset Created Successfully (1000 Rows)")
df.head()
```
